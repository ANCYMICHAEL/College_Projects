```python
new_data['Profitability'] = new_data['Revenue (Millions)']/new_data['Runtime (Minutes)']
# new_data

profit=new_data.sort_values(by='Profitability', ascending =False).head(10)
profit[['Movie Title','Profitability']]
```

```python
# new_data['Quality'] = new_data['Rating'].apply(
#     lambda x: 'Poor' if x < 6 else ('Average' if x < 7.5 else 'Excellent')
# )
# new_data


def categorize_rating(Rating):
    if Rating < 6:
        return 'Poor'
    elif Rating < 7.5:
        return 'Average'
    else:
        return 'Excellent'

new_data['Quality'] = new_data['Rating'].apply(categorize_rating)

new_data
```
